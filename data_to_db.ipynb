## Librairies


```python
import pandas as pd
import numpy as np
import re
import requests
import sqlite3
```

```python
# GLOBAL

# the url of the database
swapi_base_url = "https://swapi.dev/api/"

# categories of data
resources = ["people", "films", "starships", "planets", "species", "vehicles"]
```

## Extract


Saving inforamation to the DataFrame.

```python
# fetch data from SWAPI
def fetch_swapi_data(url):
    data = []
    page = 1
    no_data_page = 0 # num of pages with no data going one after another

    while True:
        response = requests.get(f"{url}{page}")
        if response.status_code == 200:
            print(f"Page {url}{page} : data found")
            data.append(response.json())
            no_data_page = 0 
        elif response.status_code == 404:
            print(f"Page {url}{page} : no data found")
            no_data_page += 1
            if no_data_page > 10: # if there are more than 10 blank pages in a categoru in a row, assume that the last page of the category
                print(f"Probably, reached the end of data for {url} on {url}{page-no_data_page}")
                break
        else:
            print(f"Page {url}{page} : something is wrong")
            print(f"Failed to retrieve SWAPI data. Status code: {response.status_code}")
        page += 1
    return data


def extract_swapi_data():
    df_swapi = {}  
    for resource in resources:
        df_swapi[resource] = pd.DataFrame.from_dict(fetch_swapi_data(swapi_base_url + resource + "/"))
        # print(f"{resource} :  {df_swapi[resource].shape[0]}")
        # display(df_swapi[resource])
    return df_swapi


df_swapi = extract_swapi_data()

for resource in resources:
        print(resource)
        display(df_swapi[resource].head(1))
```

# To dataset



## Transform


Deleting web-based information

```python
# find the index of the item
def extract_index(url):
    match = re.search(r'/(\d+)/$', url) # Use regular expression to extract the index (digits) from the URL
    if match:
        return int(match.group(1))  # Convert to integer
    else:
        return None

# change the names of the columns to indicate the connections on id
def url_to_index_keys(df):
    for resource in resources:
        df = df.rename(columns={
            resource: resource + '_id',
            'url': 'id',
            'characters' : 'characters_id',
            'residents' : 'residents_id',
            'pilots' : 'pilots_id',
            'homeworld' : 'homeworld_id'
            })
    return df

# change the url link to the index
def url_to_index_values(df):
    for col in df.columns:
        df[col] = df[col].apply(lambda cell: [extract_index(item) if isinstance(item, str) and re.search(r'/(\d+)/$', item)
                                            else item for item in cell] if isinstance(cell, list) 
                                            else extract_index(cell) if isinstance(cell, str) and re.search(r'/(\d+)/$', cell) 
                                            else cell)
    return df

def url_to_index(df):
    df = url_to_index_keys(df)
    df = url_to_index_values(df)
    return df

# delete info about web pages
def delete_online_info(df):
    if 'created' in df.columns:
        df = df.drop(columns=['created'])
    if 'edited' in df.columns:
        df = df.drop(columns=['edited'])
    return df
    

df_swapi_cleaned = df_swapi.copy()
for resource in resources:
    df_swapi_cleaned[resource] = url_to_index(df_swapi_cleaned[resource])
    df_swapi_cleaned[resource] = delete_online_info(df_swapi_cleaned[resource])
    # print(df_swapi_cleaned[resource].head(1))



```

Find lists in the DataFrame

```python
# check if the string can be classified as a sentence instead of the list
def check_if_sentence(value):
    # Check if a string value looks like a sentence
    return isinstance(value, str) and (('\n' in value) or value.endswith(('.', '!', '?', '...')))

# split the strings into lists
def transform_comma_separated_string(value):
    # Split a comma-separated string into a list
    return value.split(', ')

#  check if item is a list
def find_list(value):
    return ', ' in value and not check_if_sentence(value)

#  transworm strings to lists
def transform_dataframe_strings_to_lists(df):
    for column in df.columns:
        if df[column].dtype == object:  # Check if the column contains strings
            # check if there are any lists and only lists in the column 
            sentences_found = any(df[column].apply(check_if_sentence)) # if sentence is in a column
            list_found = any(df[column].apply(find_list)) # if list is in a column
            
            if list_found and not sentences_found: # if there are lists in a column, change column type to the list 
                df[column] = df[column].apply(transform_comma_separated_string) 
    return df


for resource in resources:
    df_swapi_cleaned[resource] = transform_dataframe_strings_to_lists(df_swapi_cleaned[resource])
    display(df_swapi_cleaned[resource].head(1))
```

Since SQL cannot handle lists, we create extra DataFrames, where the connections will be indicated.

```python
# transwer lists to seperate dataframes
def seperate_lists(df):
    df_exploded = {}
    for column in df.columns:
        if isinstance(df[column].iloc[0], list):  # Check if cell in the column contains a list
            df_exploded[column[:-3] if '_id' in column else column] = df[['id', column]].explode(column).dropna()
            df = df.drop(column, axis=1)
    return(df, df_exploded)

df_connections = {}
for resource in resources:
    df_swapi_cleaned[resource], df_connections[resource] = seperate_lists(df_swapi_cleaned[resource])
    for col in df_connections[resource]:
        display(df_connections[resource][col].head(3))
    
for resource in resources:
    # print(f"{resource} :  {df_swapi_cleaned[resource].shape[0]}")
    print(f"{resource}")
    display(df_swapi_cleaned[resource].head(1))
```

Turn data to the fitting type

```python
# change data type
def convert_object_columns(df):
    for column in df.columns:
        if df[column].dtype == 'object':
            try:
                # Try to convert the column to numerical
                df[column] = pd.to_numeric(df[column])
            except ValueError:
                try:
                    # Try to convert the column to a date
                    df[column] = pd.to_datetime(df[column])
                except ValueError:
                    # Handle any conversion errors or NaT (Not-a-Time) values
                    pass
    return df

for resource in resources:
    df_swapi_cleaned[resource] = convert_object_columns(df_swapi_cleaned[resource])
    for resource2 in df_connections[resource]:
        df_connections[resource][resource2] = convert_object_columns(df_connections[resource][resource2])
# print(df_connections)
```

Show data

```python
for resource1 in df_connections:
    for resource2 in df_connections[resource1]:
        print(resource1  + '_to_'+ resource2)
        print(df_connections[resource1][resource2].dtypes)
        display(df_connections[resource1][resource2].head(1))
for resource in resources:
    print(resource)
    print(df_swapi_cleaned[resource].dtypes)
    display(df_swapi_cleaned[resource].head())
```

## Load


From the DataFrames, create an SQL databese.

```python
conn = sqlite3.connect('swapi.db')
c = conn.cursor()

for resource in resources:
    c.execute(f"CREATE TABLE IF NOT EXISTS {resource} ({', '.join(df_swapi_cleaned[resource].columns)})")
    print(resource)
    conn.commit()
    df_swapi_cleaned[resource].to_sql(resource, conn, if_exists='replace', index=False)

for resource1 in df_connections:
    for resource2 in df_connections[resource1]:
        print(resource1, resource2)
        table_name = f"{resource1}_to_{resource2}"
        column_names = [f"col_{column}" for column in df_connections[resource1][resource2].columns]
        c.execute(f"CREATE TABLE IF NOT EXISTS {table_name} ({', '.join(column_names)})")
        conn.commit()
        df_connections[resource1][resource2].to_sql(table_name, conn, if_exists='replace', index=False)

# Close the database connection
conn.close()
```
